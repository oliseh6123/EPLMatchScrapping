### **Cleaning our data for machine learning**

converting date to datetime to make for best numerical(countable) value for a better machine learning modeling. 

```python
matches["date"] = pd.to_datetime(matches["date"])
```

```python
matches.dtypes
```

### **Creating predictors for machine learning**

In other to have a good predictor, we need to consider venue column which shows team is at home and away which has a considerable great effect on the outcome of the games. 

note: we are taking this measures because we need to use 

we will categories it to 0s - 1s

*   Home = 1

*   Away = 0

Also we will categories the oppose teams with numerical values assign

start from 0

Also we will be considering the hours at which the matches were played as sometimes(with historical data) teams tend to do better playing at a certain time.

Also we will categories the day fo the week.

*   monday = 0
*   tuesday = 1 etc

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes
# matches
```

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
# matches
```

```python
matches["hour"] = matches["time"].str.replace(":.+","", regex=True).astype("int")
#  matches
```

```python
matches["day_code"] = matches["date"].dt.dayofweek
# matches
```

Setting up a target if a team wins or not

```python
matches["target"] = (matches["result"] == "W").astype("int")
#matches
```

**Let start building a machine learning model**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train = matches[matches["date"] < "2022-01-01"]
```

```python
test = matches[matches["date"] > "2022-01-01"]
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code"]
```

```python
rf.fit(train[predictors], train["target"])
```

**Now we can run predictions using the test data of the model**

```python
preds = rf.predict(test[predictors])
```

**Improve the accuracy of the model**

predicting the percentage of the time the team actually won or lose

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["target"], preds)
```

We are going to be using the previous games to predict the future

```python
group
```

```python
def rolling_averages(group, cols, new_cols):
  group = group.sort_values("date")
  rolling_stats = group[cols].rolling(3, closed="left").mean()
  group[new_cols] = rolling_stats
  group = group.dropna(subset=new_cols)
  return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

**Retraining our machine learning model**

```python
def make_predictions(data, predictors):
  train = data[data["date"] < "2022-01-01"]
  test = data[data["date"] > "2022-01-01"]
  rf.fit(train[predictors], train["target"])
  preds = rf.predict(test[predictors])
  combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
  precision = precision_score(test["target"], preds)
  return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```
